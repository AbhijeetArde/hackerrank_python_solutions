```python
# Better Solution
def isValid1(s):
    c = Counter(s)
    print(c)
    freq = Counter(c.values())
    print(freq)
    if len(freq) == 1:
        return "YES"
    elif len(freq) == 2:
        key_max = max(freq.keys())
        print(key_max)
        key_min = min(freq.keys())
        print(key_min)
        if key_max - key_min == 1 and freq[key_max] == 1:
            return "YES"
        elif key_min == 1 and freq[key_min] == 1:
            return "YES"
    return "NO"
```
