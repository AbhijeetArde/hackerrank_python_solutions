```python
def sub_value(arr):
    new_arr = []
    for i in range(len(arr)-1):
        if arr[i] > max(arr[i+1:]):
            new_arr.append(arr[i])
    new_arr.append(arr[i+1])
    print(new_arr)
```

```python
sub_value([20,17,19,18,12,16,10,4,6,3])
```

```python
sub_value([25,30,21,22,14,10,5,26])
```
