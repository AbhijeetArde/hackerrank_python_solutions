```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the separateNumbers function below.
def separateNumbers(s):
    if s[0] == s:
        print('No')
        return
    for i in range(1, len(s)):
        mystack = []
        mystack.append(s[:i])
        print(mystack)
        while len(''.join(mystack)) < len(s):
            mystack.append(str(int(mystack[-1]) + 1))
        if ''.join(mystack) == s:
            print('Yes', mystack[0])
            break
        if i == len(s) - 1:
            print('No')
        
if __name__ == '__main__':
    q = int(input())

    for q_itr in range(q):
        s = input()

        separateNumbers(s)
```

```python
def separateNumbers(s):
    n = len(s)
    i=1
    while i<=n//2:
        j=i
        a = int(s[:i])
        flag =1
        while j<n:
            b = str(a+1)
            if s[j:j+len(b)]!=b:
                flag =0
                break
            j+=len(b)
            a = int(b)
        if flag:
            print('YES '+s[:i])
            return
        i+=1
    print('NO')
```
